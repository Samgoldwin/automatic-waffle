Develop a program to create histograms for all numerical features and
analyze the distribution of each feature. Generate box plots for all numerical
features and identify any outliers. Use California Housing dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
```

```python
california_housing = fetch_california_housing()
data = pd.DataFrame(california_housing.data,columns=california_housing.feature_names)
```

```python
data['MedHouseVal'] = california_housing.target
```

```python
print(data['MedHouseVal'])
```

```python
data.hist(bins=30,figsize=(15,10))
plt.suptitle("histogram of all numerical features")
plt.show()
```

```python
plt.figure(figsize=(15,10))
```

```python
for i,column in enumerate(data.columns):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=data[column])
    plt.suptitle(column)
plt.tight_layout()
```

```python
Q1= data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
```

```python
outlier = (( data < (Q1-1.5*IQR))|(data> (Q3+1.5*IQR)).sum())
print(outlier)
```
