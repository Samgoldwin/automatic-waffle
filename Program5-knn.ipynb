```python
import numpy as np
import matplotlib
matplotlib.use('TkAgg') # Use the TkAgg backend for stable display
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
xval  = np.random.rand(100,1)
```

```python
print(xval)
```

```python
y_labels = np.array(['Class1' if x <= 0.5 else 'Class2' for x in xval.flatten()])
```

```python
X_train = xval[:50] # First 50 points
y_train = y_labels[:50] # First 50 labels
X_test = xval[50:] # Remaining 50 points
y_test = y_labels[50:] # Remaining 50 labels
```

```python
k_values = [1, 2, 3, 4, 5, 20, 30]
plt.figure(figsize=(12, 8))
```

```python
for i, k in enumerate(k_values, 1):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy = knn.score(X_test, y_test)
    print(f"Accuracy for k={k}:{accuracy:.2f}")
# Predict the labels for the test set
    y_pred = knn.predict(X_test)
# Plot the decision boundary and the points
    plt.subplot(3, 3, i)
    plt.scatter(X_test, y_test, color='blue', label='True Label')
    plt.scatter(X_test, y_pred, color='red',marker = 'x', label='Predicted Label')
    plt.title(f"KNN withk={k}")
    plt.xlabel("X value")
    plt.ylabel("Class Label") 
    plt.legend(loc='best')
    plt.grid(True)
    plt.tight_layout()
plt.show()
```

* Randomly generating values learnt with seed value.
* Understanding about k nearest trees and with impoartance of k value .
* Understood visualising the scatter with points and markers
