Develop a program to implement k-means clustering using Wisconsin
Breast Cancer data set and visualize the clustering result

```python
#loading dependency
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = load_breast_cancer()
X= data.data
y= data.target
```

```python
scale = StandardScaler()
x_scale = scale.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=2,random_state=42)
y_pred = kmeans.fit_predict(x_scale)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x_scale[:,0],x_scale[:,1],c=y_pred,cmap='viridis',edgecolors='k')
plt.title("cluster on breast cancer data")
plt.colorbar(label="Cluster")
```

```python
print("Cluster centers:\n", kmeans.cluster_centers_)
```
