```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
# Standardize the features (important for PCA)
Scaler = StandardScaler()
```

```python
Scaled = Scaler.fit_transform(data)
```

```python
pca = PCA(n_components=2)
```

```python
pcafit = pca.fit_transform(Scaled)
```

```python
pcadf = pd.DataFrame(pcafit,columns=['PC1','PC2'])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pcadf['PC1'],pcadf['PC2'],c=iris.target,cmap='viridis')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component2")
plt.colorbar(label='Target')
plt.show()
```

* Learnt about dimensionality reduction using methods such as PCA
* how data can preserve quality while not storing large irrelevant data( quantity)
* Learnt how scaling methods like min-max and Standard scaling methods are done to scale the data to its variance.
